```python
import pandas
```

```python
import spacy
```

```python
# initialize language model
nlp = spacy.load("en_core_web_md")

# add pipeline (declared through entry_points in setup.py)
linker = nlp.add_pipe("entityLinker", last=True)
```

```python
system_census = pandas.read_excel('FY22 System Census - API AI Software Products_2022_0128.xlsx', 
                                  sheet_name="System Software Products")
```

```python
entities = []
for mention in system_census['Product Name'].value_counts().index:
    doc = nlp(mention)
    if doc._.linkedEntities:
        for ent in doc._.linkedEntities:
            entities.append({
                'Produc Name': mention,
                'wd_id': f'Q{ent.identifier}',
                'wd_label': ent.label,
                'wd_desc': ent.description
            })
            break
    else:
        entities.append({
            'Produc Name': mention,
        })

    
```

```python
pandas.DataFrame(entities).head(50)
```
