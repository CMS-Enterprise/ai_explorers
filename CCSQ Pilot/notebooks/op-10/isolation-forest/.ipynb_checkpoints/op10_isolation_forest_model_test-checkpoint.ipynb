```python
import pandas as pd
import numpy as np
import collections
from sklearn.ensemble import IsolationForest
import plotnine
from plotnine import *
```

```python
full_dataset
```

```python
full_dataset = pd.read_csv("../data/OP-10-no-backfill.csv")
modeling_dataset = full_dataset.drop(columns=['provider_id','lag1', 'lag2', 'prov_mean'])
modeling_dataset = modeling_dataset.dropna()
y_quarter_list = modeling_dataset['year'].astype(str) + 'Q' + modeling_dataset['quarter'].astype(str)
scores_list = modeling_dataset['score']
modeling_dataset = modeling_dataset.drop(columns = ['score','year', 'quarter'])
modeling_dataset
```

```python
#Isolation Forest Model
rng = np.random.RandomState(42)
if_model = IsolationForest(
    random_state = rng,
    max_samples = 128,
    n_jobs = -1)
```

```python
if_model.fit(modeling_dataset)
```

```python
results = if_model.predict(modeling_dataset)
```

```python
# -1 is Anomaly, 1 is normal

pd.DataFrame(results).value_counts()
```

```python
anomaly_scores = if_model.score_samples(modeling_dataset) * -1
```

```python
pd.DataFrame(anomaly_scores > 0.5).value_counts()
```

```python
results_dataset = modeling_dataset.copy()
results_dataset
```

```python
anomaly_scores = if_model.score_samples(modeling_dataset) * -1
```

```python
results_dataset
```

```python
results_dataset['anomaly_score'] = anomaly_scores
results_dataset['pred'] = results
results_dataset['outlier'] = (results_dataset['pred'] == -1)
results_dataset['y_quarter'] = y_quarter_list
results_dataset['score'] = scores_list
```

```python
# Detailed Outlier Analysis 
results_dataset = qs.copy()
threshold_range = np.arange(0.55, 0.675, 0.025)
threshold_range = np.insert(0.01, 0, threshold_range)
for t in threshold_range:
    results_dataset['>' + str(round(t,3))] = results_dataset['anomaly_score'] > t
results_dataset = results_dataset.rename(columns = {">0.01": ">0.01 (All)"})
values = [">" + str(round(s,3)) for s in np.arange(0.55, 0.675, 0.025)]
values.insert(0,">0.01 (All)")
results_dataset = pd.melt(
    results_dataset,
    id_vars = results_dataset.columns.difference(values),
    value_vars = values,
    var_name = 'threshold',
    value_name = 'anomaly'
)
results_dataset
```

```python
anamolies = results_dataset[results_dataset['anomaly'] == True]
```

```python
(ggplot(anamolies,    
        aes(x = 'score',
            color = 'threshold',
            fill = 'threshold',
            group = 'threshold')) +
    xlab("SEP-01 Score") +
    ylab("Count") +
    ggtitle("Anomaly Score Distribution By Threshold") +
    labs(color = "Threshold", fill = "Threshold") +
    scale_color_manual(values = ["#7d7c7a","#b1c7b3","#f1eac8","#e5b9ad","#d98994","#d0587e"]) +
    scale_fill_manual(values = ["#7d7c7a","#b1c7b3","#f1eac8","#e5b9ad","#d98994","#d0587e"]) +
    geom_histogram(bins = 100, position = "nudge")
    
)
```

```python
(ggplot(anamolies[anamolies['threshold'] != ">0.01 (All)"],    
        aes(x = 'lag_diff',
            y = 'prov_mean_diff',
            color = 'threshold',
            fill = 'threshold',
            group = 'threshold')) +
    xlab("Lag Difference") +
    ylab("Provider Mean Difference") +
    ggtitle("Anomaly Score Distribution By Threshold") +
    labs(color = "Threshold", fill = "Threshold") +
    scale_color_manual(values = ["#b1c7b3","#f1eac8","#e5b9ad","#d98994","#d0587e"]) +
    scale_fill_manual(values = ["#b1c7b3","#f1eac8","#e5b9ad","#d98994","#d0587e"]) +
    geom_point() +
    facet_wrap('~y_quarter')
    
)
```
