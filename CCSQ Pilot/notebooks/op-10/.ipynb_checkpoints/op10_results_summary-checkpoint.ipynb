```python
import pandas as pd
import numpy as np
import collections
import plotnine
from plotnine import *
import os
import ast
import random
import pickle
```

```python
op10_filtered_path = "./data/OP-10-no-backfill.csv"
op10_dataset = pd.read_csv(op10_filtered_path)
op10_dataset = op10_dataset.dropna(subset = ['lag1'])
op10_dataset['y_quarter'] = op10_dataset['year'].astype(str) + 'Q' + op10_dataset['quarter'].astype(str)
op10_dataset
```

```python
xgbr_model_path = './xgboost-regression/xgbr_model_final.pkl'
iforest_model_path = './isolation-forest/iforest_model_final.pkl'
```

```python
def load_and_predict_xgbr(model_file, modeling_dataset, target, threshold):
    xgbr_model = pickle.load(open(model_file, 'rb'))
    scores = xgbr_model.predict(modeling_dataset)
    prediction = abs(target - scores) > threshold
    
    return scores, prediction
```

```python
def load_and_predict_iforest(model_file, modeling_dataset, threshold): 
    if_model = pickle.load(open(model_file, 'rb'))
    scores = if_model.score_samples(modeling_dataset) * -1
    prediction = scores > threshold

    return scores, prediction
```

```python
xgbr_scores, xgbr_prediction =\
    load_and_predict_xgbr(xgbr_model_path,
                          op10_dataset[['lag1','prov_mean']],
                          op10_dataset['score'],
                          4)
```

```python
iforest_scores, iforest_prediction =\
    load_and_predict_iforest(iforest_model_path,
                          op10_dataset[['lag_diff','prov_mean_diff']],
                          .5)
```

```python
xgb_op10_dataset = op10_dataset.copy()[['score','y_quarter','provider_id','prov_mean','lag_diff','prov_mean_diff']]
xgb_op10_dataset['outlier'] = xgbr_prediction
xgb_op10_dataset['model_score'] = xgbr_scores
xgb_op10_dataset['model'] = "XGBoost Regression"
xgb_op10_dataset
```

```python
if_op10_dataset = op10_dataset.copy()[['score', 'y_quarter','provider_id','prov_mean','lag_diff','prov_mean_diff']]
if_op10_dataset['outlier'] = iforest_prediction
if_op10_dataset['model_score'] = iforest_scores
if_op10_dataset['model'] = "Isolation Forest"
if_op10_dataset
```

```python
comparison_df = pd.concat([xgb_op10_dataset, if_op10_dataset], axis = 0)
comparison_df
```

# Modeling Process

### Model Training and Selection
The XGBoost Regression model utilizes lag, year, and quarter to predict the observed measure score (supervised).

The Isolation Forest utilizes score, lag, year, and quarter to generate an anomaly score (unsupervised).

Both models were tuned under a range of hyperparemeters via a grid search. Final candidate models were selected by either 5-fold cross validated RMSE (for XGBoost) or total AUROC (for Isolation Forest, described further below)

### Artifical Anomaly Test
The artificial anomaly test is conducted as follows:

1.	A list of “shift values” 2 to 4 times the standard deviations of the measure data is made (for SEP-01, 36 through 72)
2.	For each shift value, 20k observations are bootstrapped sampled (sampled with replacement) to form a new dataset.
3.	Half of the observations in each shift value dataset are given a mock score by adding or subtracting the shift value from the actual score. These observations are labeled as anomalies.
4.	The generated datasets for each shift value are combined into one dataset.
5.	The chosen model predicts whether each observation is an anomaly.
6.	Accuracy is judged by AUROC (area under the receiver operator characteristic curve), a commonly used metric to evaluate performance of categorical models based on the true positive and false positive rates.

### Outlier thresholds

Two types of thresholds are determined during the final step of the model selection process: 

**XGBoost Regression:** Predicts the observation as an anomaly if the difference between the actual measure score and the XGBoost predicted measure score is greater than the threshold.

**Isolation Forest:** Predicts the observation as an anomaly if the isolation forest anomaly score is greater than the threshold.

In both cases, the thresholds were determined utilizing the artificial anomaly test. AUROC is calculated for each potential threshold value, and the threshold that produces the best results was selected. For SEP-01, the threshold was 15 for the XGBoost Regression model, and 0.59 for the Isolation Forest model.

# XGBoost Regression Model

#### In-sample RMSE and MAE (lower is better)

```python
rmse = np.sqrt(np.mean((xgb_op10_dataset['score'] - xgb_op10_dataset['model_score'])**2))
mae = np.sqrt(np.mean(abs(xgb_op10_dataset['score'] - xgb_op10_dataset['model_score'])))

pd.DataFrame({"RMSE":rmse, "MAE":mae}, index = [0])
```

#### Threshold Testing and Performance (higher is better)
A threshold of 3 performs best, with an AUROC of 0.841. A threshold of 4 was selected as this makes model is more conservative and affects the AUROC marginally with a score of 0.832

# Isolation Forest Model

### Threshold Test and Performance (higher is better)
A threshold of .5 performs best, with an AUROC of 0.773

```python
display(comparison_df[['outlier', 'score','model']]
     .rename(columns = {'outlier':'Predicted Outlier',
                        'score':'Score',
                        'model':'Model'})
     .groupby(['Predicted Outlier', 'Model'])
     .describe(percentiles = [.1, .25, .5, .75, .9])
     .transpose()
    .round(1)
)
```

#### XGBoost Outlier Counts by Quarter

```python
xgb = comparison_df.query('model == "XGBoost Regression"')

summary_table = pd.DataFrame({"outlier count": xgb.loc[ xgb['outlier'] == True, 'y_quarter'].value_counts(),
                              "total": xgb['y_quarter'].value_counts()})
summary_table['% outlier'] = summary_table["outlier count"] / summary_table["total"]
summary_table['% outlier'] = round(summary_table['% outlier']*100,2)
summary_table
```

#### Isolation Forest Outlier Counts by Quarter

```python
iforest = comparison_df.query('model == "Isolation Forest"')

summary_table = pd.DataFrame({"outlier count": iforest.loc[ iforest['outlier'] == True, 'y_quarter'].value_counts(),
                              "total": iforest['y_quarter'].value_counts()})
summary_table['% outlier'] = summary_table["outlier count"] / summary_table["total"]
summary_table['% outlier'] = round(summary_table['% outlier']*100,2)
summary_table
```
