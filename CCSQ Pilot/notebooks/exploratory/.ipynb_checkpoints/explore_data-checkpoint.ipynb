```python
import pandas as pd
import boto3
import altair as alt
from plotnine import *
```

```python
s3 = boto3.client('s3')

bucket_name = 'haip-dev-training-dataset'
s3_oie_dataset_path = 'oie_historical_dataset.csv'
s3_oie_dataset_dates_path = 'oie_with_dates.csv'

local_oie_dataset_path = '../tmp/oie_historical_dataset.csv'
local_oie_dataset_dates_path = '../tmp/oie_with_dates.csv'
```

```python
s3.download_file(bucket_name, s3_oie_dataset_path, local_oie_dataset_path)
s3.download_file(bucket_name, s3_oie_dataset_dates_path, local_oie_dataset_dates_path)
```

```python
oie_dataset = pd.read_csv(local_oie_dataset_path)
oie_dataset_dates = pd.read_csv(local_oie_dataset_dates_path)
```

```python
oie_dataset_dates
```

oie_dataset5k sample of data

```python
plot_data = oie_dataset_dates\
    .loc[oie_dataset_dates['providerId'] < 10100]
#    .loc[oie_dataset_dates['measureId'] == 'OP_23']



display(plot_data)

plot = (
    ggplot(plot_data, aes('year','score', color = 'factor(providerId)'))
    + geom_line(alpha = 0.4, show_legend = False)
    + facet_wrap('~measureId')
 )

display(plot)
```
