```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import collections
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from sklearn.metrics import roc_auc_score
import plotnine
from plotnine import *
import os
import pickle
```

```python
modeling_filtered_path = "../../data/OP-22.csv"
modeling_dataset = pd.read_csv(modeling_filtered_path)
modeling_dataset = modeling_dataset.drop(columns = ['lag2']).dropna()
modeling_dataset
```

```python
def artificial_anomaly_dist(model_data, 
                                threshold_range = range(1,40),
                                shift_range = range(1,75),
                                shift_n = 1000,  
                                scatter = False,
                                rng = 42):
    
    shift_values = np.repeat(shift_range, shift_n)
    bootstrap_sample = model_data.sample(len(shift_values), 
                                         replace = True,
                                         random_state = rng)
    bootstrap_sample['shift'] = shift_values
    bootstrap_sample['mock_score'] = (
        bootstrap_sample['score']
        + bootstrap_sample['shift']
        * np.random.choice([-1,1],len(bootstrap_sample))
        * np.random.uniform(0,1)
    )
    if scatter == False:
        bootstrap_sample.loc[bootstrap_sample['mock_score'] < 0, 'mock_score'] = 0
        bootstrap_sample.loc[bootstrap_sample['mock_score'] > 100, 'mock_score'] = 100
        
    elif scatter == True:
        bootstrap_sample.loc[bootstrap_sample['mock_score'] > 100, 'mock_score'] = (
            bootstrap_sample.loc[bootstrap_sample['mock_score'] > 100, 'score']
            - bootstrap_sample.loc[bootstrap_sample['mock_score'] > 100, 'shift']
        )
        bootstrap_sample.loc[bootstrap_sample['mock_score'] < 0,'mock_score'] = (
             bootstrap_sample.loc[bootstrap_sample['mock_score'] < 0, 'score']
             + bootstrap_sample.loc[bootstrap_sample['mock_score'] < 0, 'shift']
         )
        bootstrap_sample.loc[bootstrap_sample['mock_score'] < 0, 'mock_score'] = 0
        bootstrap_sample.loc[bootstrap_sample['mock_score'] > 100, 'mock_score'] = 100

    return bootstrap_sample
```

```python
l = []
l.extend(range(1,16,1))
l
```

```python
dist_results_scatter = artificial_anomaly_dist(modeling_dataset, shift_range = l, scatter = True)
dist_results_scatter = dist_results_scatter[['provider_id', 'score','mock_score', 'shift']]
dist_results_scatter
```

```python
melted_dist_scatter = pd.melt(dist_results_scatter, id_vars = ['provider_id','shift'])
melted_dist_scatter.loc[melted_dist_scatter['variable'] == 'score', 'variable'] = "Original Bootstrap Sample"
melted_dist_scatter.loc[melted_dist_scatter['variable'] == 'mock_score', 'variable'] = "Artificial Anomaly"
melted_dist_scatter = melted_dist_scatter.rename(columns = {'variable': 'group','value': 'score' })
melted_dist_scatter
```

```python
plotnine.options.figure_size = (12,24)
graph_df = melted_dist_scatter.copy()
graph_df['shift'] = ('Shift ' + graph_df['shift'].astype(str).str.zfill(2))
(ggplot(graph_df, 
        aes(x = 'score',
            color = 'group',
            fill = 'group')
       )
 + labs(x = 'OP-22 Score', 
        y = 'Count',
        color = 'Group',
        fill = 'Group')
 + geom_histogram(binwidth = 1, alpha = 0.7) 
 + scale_color_manual(values = ["#F8766D", "#00BFC4"], 
                      guide = False)
 + scale_fill_manual(values = ["#F8766D", "#00BFC4"],
                     guide = False)
 + facet_grid('shift ~ group ')
)
```
